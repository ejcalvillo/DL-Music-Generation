```python
import pretty_midi
import pandas as pd
import numpy as np
```

```python
def midi_to_notes(midi_file: str) -> pd.DataFrame:
    pm = pretty_midi.PrettyMIDI(midi_file)
    instrument = pm.instruments[0] # MAESTRO is solo piano
    notes = []

    # Sort notes by start time
    sorted_notes = sorted(instrument.notes, key=lambda note: note.start)
    prev_start = sorted_notes[0].start

    for note in sorted_notes:
        start = note.start
        end = note.end
        notes.append({
            'pitch': note.pitch,
            'step': start - prev_start,
            'duration': end - start,
        })
        prev_start = start

    return pd.DataFrame(notes)
```

```python
raw_notes = midi_to_notes('./maestro/2004/MIDI-Unprocessed_SMF_02_R1_2004_01-05_ORIG_MID--AUDIO_02_R1_2004_05_Track05_wav.midi')
```
